# Movie Reviews

```python
import pandas as pd

data = pd.read_csv("reviews.csv")

data.head()
```

The dataset is made up of positive and negative movie reviews.

## Preprocessing

üëá Remove punctuation and lower case the text.

```python
import string

def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
        
def lower_text(text):
    return text.lower()

data.reviews = data.reviews.apply(remove_punctuation)

data.reviews = data.reviews.apply(lower_text)

data.head()
```

## Bag-of-Words modelling

üëá Using `cross_validate`, score a Multinomial Naive Bayes model trained on a Bag-of-Word representation of the texts.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(data.reviews)

X.toarray()
```

```python
X_bow = pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names())

X_bow
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

y = data.target

model.fit(X_bow, y)

cross_val_score(model, X_bow, y).mean()
```

## N-gram modelling

üëá Using `cross_validate`, score a Multinomial Naive Bayes model trained on a 2-gram Bag-of-Word representation of the texts.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vectorizer = TfidfVectorizer(ngram_range = (2,2))

X2 = tf_idf_vectorizer.fit_transform(data.reviews)

X2.toarray()

X2 = pd.DataFrame(X2.toarray(),columns = tf_idf_vectorizer.get_feature_names())

X2
```

```python
cross_val_score(model, X2, y).mean()
```
