```python
import pandas as pd

df = pd.read_csv("https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/reviews.csv")

df.head()
```

```python
df.review_score[28884] = '5'
df.review_score[58381] = '5'
```

```python
df['review_score'] = pd.to_numeric(df['review_score'])
```

üëá Analyse the comment reviews to understand what causes the bad review scores.

```python
df = df.loc[df['review_score'] < 4]
df
```

```python
import numpy as np

df.review_comment_message = df.review_comment_message.replace(np.nan, '')
```

```python
import string 

def remove_numbers(text):
    return ''.join(word for word in text if not word.isdigit())

def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

def lower_case(text):
    return text.lower()
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('portuguese')) 

def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    return [w for w in word_tokens if not w in stop_words]
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(word) for word in text]

def list_to_string(list):
    return ' '.join(list)
```

```python
df.review_comment_message = df.review_comment_message.apply(remove_punctuation)
df.review_comment_message = df.review_comment_message.apply(remove_numbers)
df.review_comment_message = df.review_comment_message.apply(lower_case)
df.review_comment_message = df.review_comment_message.apply(remove_stopwords)
df.review_comment_message = df.review_comment_message.apply(lemmatize_text)
df.review_comment_message = df.review_comment_message.apply(list_to_string)

df
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer().fit(df.review_comment_message)

data_vectorized = vectorizer.transform(df.review_comment_message)

lda_model = LatentDirichletAllocation(n_components=4).fit(data_vectorized)

def print_topics(model, vectorizer):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-10 - 1:-1]])
```

```python
print_topics(lda_model, vectorizer)
```
